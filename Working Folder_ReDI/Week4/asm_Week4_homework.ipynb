#### How many vegan recipes are there (Vegan = Vegetarian and dairy free)?

```python
mask1 = df['vegetarian'] == True
mask2 = df["dairyFree"] == True
df_vegan = df[mask1 & mask2]
# print(df1)
print(f"Number of vegan recipes: {len(df_vegan)}")
```

#### Compare the average amount of Vitamin B12 for the vegan and non-vegan recipes. How reliable are the results?

```python
mask1 = df['vegetarian'] == True
mask2 = df["dairyFree"] == True
df_vegan = df[mask1 & mask2]
df_nonvegan = df[mask1 | mask2]
print('Vegan Mean for Vit B12:')
print(df_vegan['VitaminB12percentOfDailyNeed'].mean())
print('Non-Vegan Mean for Vit B12:')
print(df_nonvegan['VitaminB12percentOfDailyNeed'].mean())
```

#### Compare the average health Score for both types of recipes (vegan vs non-vegan).

```python
print("Average Health Score of Vegan recipes :")
print(df_vegan['healthScore'].mean())
print("Average Health Score of Non-Vegan recipes :")
print(df_nonvegan['healthScore'].mean())
```
