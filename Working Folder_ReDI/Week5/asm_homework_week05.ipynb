#### a) Import the dataset into a pandas DataFrame. Convert the time column into a pandas.Timestamp object. 

```python
import pandas as pd
df = pd.read_csv("data/temperature_signals.csv")
df["time"] = pd.to_datetime(df.time, format="%Y-%m-%d %H:%M:%S")
df.info()
```

#### c) Now visualise each of the 3 temperature signals.
- First, you want to create a figure with 3 different axes instances, one axis instance for each temperature signal
- Then you visualize each signal. Do you use a scatter or line plot? Try to justify your choice with a couple of words
- Because you want to discuss your experiment with your colleagues, you also want to make the figure pretty. Label each of the axes with information that you consider important

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16,5))

ax1 = fig.add_subplot(2,2,1) 
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3) 

ax1.plot(df.index, df["temp1"], label="signal1", color="red")
ax2.plot(df.index, df["temp2"], label="signal2", color="yellow")
ax3.plot(df.index, df["temp3"], label="signal3", color="green")

ax1.set_title("Axes instance for plot of Temp1 against Timestamp")
ax2.set_title("Axes instance for plot of Temp2 against Timestamp")
ax3.set_title("Axes instance for plot of Temp3 against Timestamp")

fig = plt.gcf()
plt.tight_layout()

print("A line plot with different colors gives a clear picture of the temparature trends for the 3 columns of the dataframe. It is easy to compare each line against the other to analyze further.")
```

#### e) Now create one figure with one axis element to visualise the temperature of signal 1 and 2 in one coordinate system. Don't forget to label your x- and y-axis

```python
fig, ax = plt.subplots(figsize=(150, 10))
ax.plot(df.index, df["temp1"], label="signal1", color="red")
ax.plot(df.index, df["temp2"], label="signal2", color="blue")

ax.set_title('2 Temperature Signals')
ax.set_xlabel('Timestamp')
ax.set_ylabel('Signals')

ax.set_xticks(df.index)
ax.set_xticklabels(df['time'], rotation=90)
ax.legend()
ax.grid(which="both")
plt.tight_layout()
```

#### f) Describe in 1-2 sentences what you now see
